## Negative Values (no irrigation) for all years combined

```python
## List of all years
years = ['2016', '2017', '2018', '2019', '2020', '2021']

# Create an empty list to store the percentage DataFrames for each year
percentage_dfs = []

for year in years:
    # Read the CSV file for the current year
    csv_file = 'G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Section_IWU_{}.csv'.format(year)
    data = pd.read_csv(csv_file)

    # Select the columns of interest
    columns = data[['DISALEXI_DAYMET', 'EEMETRIC_DAYMET', 'GEESEBAL_DAYMET', 'PTJPL_DAYMET', 'SIMS_DAYMET', 'SSEBOP_DAYMET',
                'DISALEXI_gridMET', 'EEMETRIC_gridMET', 'GEESEBAL_gridMET', 'PTJPL_gridMET', 'SIMS_gridMET', 'SSEBOP_gridMET',
                'DISALEXI_IMERG', 'EEMETRIC_IMERG', 'GEESEBAL_IMERG', 'PTJPL_IMERG', 'SIMS_IMERG', 'SSEBOP_IMERG',
                'DISALEXI_NLDAS', 'EEMETRIC_NLDAS', 'GEESEBAL_NLDAS', 'PTJPL_NLDAS', 'SIMS_NLDAS', 'SSEBOP_NLDAS',
                'DISALEXI_PRISM', 'EEMETRIC_PRISM', 'GEESEBAL_PRISM', 'PTJPL_PRISM', 'SIMS_PRISM', 'SSEBOP_PRISM']]

    # Calculate the percentage of negative values for each column
    percentage_negative = columns.apply(lambda x: (x < 0).mean() * 100)

    # Create a new DataFrame to store the percentage_negative values
    percentage_df = pd.DataFrame(percentage_negative, columns=['Percentage_Negative'])

    # Append the DataFrame to the list
    percentage_dfs.append(percentage_df)

# Plot the boxplots for each year's data
plt.figure(figsize=(12, 6))
plt.boxplot([df['Percentage_Negative'] for df in percentage_dfs], labels=years, showmeans=True)
plt.ylabel('Percentage (%)', fontweight='bold')
plt.xlabel('Year', fontweight='bold')
plt.title('Percentage of Negatively estimated IWU', fontweight='bold')
#plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Per Year per Model

```python
# List of columns to include in the analysis
columns = ['DISALEXI_DAYMET', 'EEMETRIC_DAYMET', 'GEESEBAL_DAYMET', 'PTJPL_DAYMET', 'SIMS_DAYMET', 'SSEBOP_DAYMET',
                        'DISALEXI_gridMET', 'EEMETRIC_gridMET', 'GEESEBAL_gridMET', 'PTJPL_gridMET', 'SIMS_gridMET', 'SSEBOP_gridMET',
                        'DISALEXI_IMERG', 'EEMETRIC_IMERG', 'GEESEBAL_IMERG', 'PTJPL_IMERG', 'SIMS_IMERG', 'SSEBOP_IMERG',
                        'DISALEXI_NLDAS', 'EEMETRIC_NLDAS', 'GEESEBAL_NLDAS', 'PTJPL_NLDAS', 'SIMS_NLDAS', 'SSEBOP_NLDAS',
                        'DISALEXI_PRISM', 'EEMETRIC_PRISM', 'GEESEBAL_PRISM', 'PTJPL_PRISM', 'SIMS_PRISM', 'SSEBOP_PRISM']

# List of years
years = ['2016', '2017', '2018', '2019', '2020', '2021']

# Define a list of colors for each group of models
model_colors = {
    'DISALEXI_': 'red',
    'EEMETRIC_': 'blue',
    'GEESEBAL_': 'black',
    'PTJPL_': 'purple',
    'SIMS_': 'orange',
    'SSEBOP_': 'green',
}

# Create a subplot for each year's data
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), sharey=True, sharex=True)
fig.suptitle('Percentage of negative IWU in each model setup', fontweight='bold', fontsize=16)

# Create a list to store the unique model group names and corresponding colors
unique_model_groups = {}

# Process data for each year and plot in the corresponding subplot
for i, year in enumerate(years):
    # Read the CSV file for the current year
    csv_file = 'G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Section_IWU_{}.csv'.format(year)
    year_data = pd.read_csv(csv_file)
    negative_counts = year_data[columns][year_data[columns] < 0].count()
    total_rows = len(year_data)
    percentage_negative = (negative_counts / total_rows) * 100
    
    # Determine the subplot position based on the index 'i'
    row = i // 3
    col = i % 3

    # Inside the loop for plotting:
    for j, column in enumerate(columns):
        # Determine the model group based on the column name
        for model_prefix, color in model_colors.items():
            if column.startswith(model_prefix):
                model_group = model_prefix.strip('_')  # Extract the model group name without underscores
                break
        
        # Add the unique model group name to the list with its corresponding color
        if model_group not in unique_model_groups:
            unique_model_groups[model_group] = color

        axes[row, col].bar(column, percentage_negative[column], color=unique_model_groups[model_group], label=model_group)

    axes[row, col].set_title(f"Year {year}", fontweight='bold')
    #axes[row, col].set_ylabel("Percentage (%)", fontweight='bold')
    
    # Set the y-label only for the subplots in the first row (when i is 0 and 3)
    if i in [0, 3]:
        axes[row, col].set_ylabel("Percentage (%)", fontweight='bold')
    axes[row, col].set_xticklabels(columns, rotation=90)
    axes[row, col].xaxis.set_visible(False)
    axes[row, col].yaxis.set_visible(True)

# Adjust the layout and spacing between subplots
plt.tight_layout()

# Create a legend with 6 entries, one for each unique model group
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=label, markersize=10, markerfacecolor=color) for label, color in unique_model_groups.items()]
fig.legend(handles=legend_handles, labels=unique_model_groups.keys(), loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```

## Spatially plot only one year

```python
# Define the file path
file_path = 'G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Section_IWU_2016.csv'
# Load your data into a pandas DataFrame
df = pd.read_csv(file_path)

# Define the columns
columns = [
    'DISALEXI_DAYMET', 'EEMETRIC_DAYMET', 'GEESEBAL_DAYMET', 'PTJPL_DAYMET', 'SIMS_DAYMET', 'SSEBOP_DAYMET',
    'DISALEXI_gridMET', 'EEMETRIC_gridMET', 'GEESEBAL_gridMET', 'PTJPL_gridMET', 'SIMS_gridMET', 'SSEBOP_gridMET',
    'DISALEXI_IMERG', 'EEMETRIC_IMERG', 'GEESEBAL_IMERG', 'PTJPL_IMERG', 'SIMS_IMERG', 'SSEBOP_IMERG',
    'DISALEXI_NLDAS', 'EEMETRIC_NLDAS', 'GEESEBAL_NLDAS', 'PTJPL_NLDAS', 'SIMS_NLDAS', 'SSEBOP_NLDAS',
    'DISALEXI_PRISM', 'EEMETRIC_PRISM', 'GEESEBAL_PRISM', 'PTJPL_PRISM', 'SIMS_PRISM', 'SSEBOP_PRISM'
]

# Function to count how many values are less than zero in the specified columns for each row
def count_less_than_zero(row):
    return sum(row[col] < 0 for col in columns)

# Apply the function across the rows and create a new column 'count_less_zero'
df['negatives'] = df.apply(count_less_than_zero, axis=1)

# Save dataframe if intended
#df.to_csv(out_dir + 'results_negative_counts_2016.csv', index=False)

shapefile_path = "G:/001Project/Output/Boundary/Field_Boundary/Irrigated_Croplands/HPA_Irr_Croplands_2017.shp"

# Read the shapefile using geopandas
shp_file = gpd.read_file(shapefile_path)


# Merge data DataFrame with shp_file GeoDataFrame based on "LN_ID" column
merged_data = shp_file.merge(df[['LN_ID', 'negatives']], left_on='LN_ID', right_on='LN_ID')

# Plotting the 'negatives' column
fig, ax = plt.subplots(figsize=(12, 10))
merged_data.plot(column='negatives', ax=ax, legend=True, cmap='Reds',
                legend_kwds={'label': "Number of non-irrigated land parcels", 'orientation': "vertical"})

# Adding title and labels (optional)
ax.set_title('Spatial Distribution of non-irrigated land parcels')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```

## Process all negative IWU for all years and save the output

```python
# Define the base path where the files are stored and the output directory for the modified files
base_path = 'G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse'

# Define the years that we have data for
years = range(2016, 2022)

# Define the columns you want to analyze
columns_to_analyze = [
    'DISALEXI_DAYMET', 'EEMETRIC_DAYMET', 'GEESEBAL_DAYMET', 'PTJPL_DAYMET', 'SIMS_DAYMET', 'SSEBOP_DAYMET',
    'DISALEXI_gridMET', 'EEMETRIC_gridMET', 'GEESEBAL_gridMET', 'PTJPL_gridMET', 'SIMS_gridMET', 'SSEBOP_gridMET',
    'DISALEXI_IMERG', 'EEMETRIC_IMERG', 'GEESEBAL_IMERG', 'PTJPL_IMERG', 'SIMS_IMERG', 'SSEBOP_IMERG',
    'DISALEXI_NLDAS', 'EEMETRIC_NLDAS', 'GEESEBAL_NLDAS', 'PTJPL_NLDAS', 'SIMS_NLDAS', 'SSEBOP_NLDAS',
    'DISALEXI_PRISM', 'EEMETRIC_PRISM', 'GEESEBAL_PRISM', 'PTJPL_PRISM', 'SIMS_PRISM', 'SSEBOP_PRISM'
]

# Function to count how many values are less than zero in the specified columns for each row
def count_less_than_zero(row):
    return sum(row[col] < 0 for col in columns_to_analyze)

# Process each file in a loop
for year in years:
    # Construct file path
    file_path = os.path.join(base_path, f'Section_IWU_{year}.csv')

    # Check if the file exists
    if not os.path.exists(file_path):
        print(f"File for {year} does not exist: {file_path}")
        continue  # Skip this year if the file doesn't exist

    # Load data
    print(f"Processing file for {year}...")
    df = pd.read_csv(file_path)

    # Apply the function across the rows and create a new column 'not_irrigated'
    df['not_irrigated'] = df.apply(count_less_than_zero, axis=1)

    # Save the modified DataFrame back to a new CSV file
    df.to_csv(f'G:/001Project/Output/Sheets/cleaned_Results/No_Irrigation/Section_IWU_{year}_not_irrigated.csv', index=False)
    print(f"File for {year} processed and saved")

print("All files processed.")
```
