## Plot Variances

```python
import pandas as pd

file_path = "G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Var_partition/W_Climate/Var_agg_climate/aggregated_by_climate_2021.csv"
df = pd.read_csv(file_path)

import matplotlib.pyplot as plt

# Define the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bars
df.set_index('CLIMATE')[['Percentage_Variance_P', 'Percentage_Variance_ET']].plot(kind='bar', stacked=True, ax=ax)

# Set the title and labels
ax.set_title('100% Stacked Bar Chart of Percentage_Variance_P vs. Percentage_Variance_ET')
ax.set_ylabel('Percentage')
ax.set_xlabel('Climate')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

base_path = "G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Var_partition/W_Climate/Var_agg_climate/aggregated_by_climate_"
years = [2016, 2017, 2018, 2019, 2020, 2021]

dfs = {}
for year in years:
    file_path = f"{base_path}{year}.csv"
    dfs[year] = pd.read_csv(file_path)

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for i, (year, df) in enumerate(dfs.items()):
    df.set_index('CLIMATE')[['Percentage_Variance_P', 'Percentage_Variance_ET']].plot(kind='bar', stacked=True, ax=axes[i], legend=(i==0), sharey=True, sharex=True)
    axes[i].set_title(f'Year: {year}')
    axes[i].set_ylabel('Percentage')
    axes[i].set_xlabel('Climate')
    axes[i].annotate(str(year), xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12, fontweight='bold', verticalalignment='top')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("G:/001Project/Output/Sheets/cleaned_Results/irrigation_wUse/Var_partition/W_Climate/Var_agg_climate/aggregated_by_climate_2021.csv")

climates = df['CLIMATE'].unique()

for climate in climates:
    sub_df = df[df['CLIMATE'] == climate]

    # Prepare the data for plotting
    plot_data = [sub_df['Percentage_Variance_P'], sub_df['Percentage_Variance_ET']]

    # Create the box plots
    fig, ax = plt.subplots()
    ax.boxplot(plot_data)
    ax.set_xticklabels(['Percentage_Variance_P', 'Percentage_Variance_ET'])
    ax.set_title(f'Box plot for climate type: {climate}')
    plt.show()

sns.set_style("whitegrid")

for climate in climates:
    sub_df = df[df['CLIMATE'] == climate]

    # Use the melt function to reshape the DataFrame for seaborn's boxplot function
    melted_df = pd.melt(sub_df, id_vars=['CLIMATE'], value_vars=['Percentage_Variance_P', 'Percentage_Variance_ET'])

    plt.figure(figsize=(8, 5))
    sns.boxplot(x="variable", y="value", data=melted_df)
    plt.title(f'Box plot for climate type: {climate}')
    plt.show()


```
